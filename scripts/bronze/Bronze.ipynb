# read data from the volumes Step 1

```python
df = (
    spark.read.option("header", True)
    .option("inferSchema", True)
    .csv("dbfs:/Volumes/workspace/bronze/raw_sources/source_crm/cust_info.csv")
)
df.display()
```

# **Write it to **Bronze****

```python
df.write.mode("overwrite").saveAsTable("bronze.crm_cust_info")
```
